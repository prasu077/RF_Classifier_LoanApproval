```python
df.drop(columns=['loan_id'], inplace=True)
```

```python

df.columns = df.columns.str.strip()
df.columns
```

```python
y = df['loan_status']
```

```python
X = df.drop(columns=['loan_status'])
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X['education'] = le.fit_transform(X['education'])
X['self_employed'] = le.fit_transform(X['self_employed'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))

```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
bagging_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    max_features=1.0,
    bootstrap=True,
    random_state=42
)

bagging_model.fit(X_train, y_train)
```

```python
y_pred_bag = bagging_model.predict(X_test)

print("Bagging Accuracy:", accuracy_score(y_test, y_pred_bag))
```

```python
from sklearn.ensemble import AdaBoostClassifier
boosting_model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)

boosting_model.fit(X_train, y_train)

```

```python
y_pred_boost = boosting_model.predict(X_test)

print("Boosting Accuracy:", accuracy_score(y_test, y_pred_boost))
```

```python
import pickle

with open("bagging_loan_model.pkl", "wb") as f:
    pickle.dump(bagging_model, f)

from sklearn.preprocessing import LabelEncoder

edu_le = LabelEncoder()
emp_le = LabelEncoder()

X['education'] = edu_le.fit_transform(X['education'])
X['self_employed'] = emp_le.fit_transform(X['self_employed'])

pickle.dump(edu_le, open("education_encoder.pkl", "wb"))
pickle.dump(emp_le, open("self_employed_encoder.pkl", "wb"))

```
